# **Problemas de Negocio que Aborda el Dataset de Ventas de Tacos**

1. fluctuaci√≥n en la Demanda:
Uno de los principales problemas de la industria gastron√≥mica, especialmente en la comida r√°pida, es la fluctuaci√≥n en la demanda. Esto puede generar desperdicios de comida o falta de insumos. El dataset puede ayudar a analizar las variaciones de demanda en diferentes d√≠as, semanas o meses, y as√≠ poder prever con m√°s precisi√≥n la cantidad de ingredientes necesarios para cada d√≠a.

2. optimizaci√≥n de Inventarios:
Sin un sistema adecuado de gesti√≥n de inventarios, los negocios pueden enfrentar p√©rdidas de productos perecederos como tortillas, carne, salsa, etc. Con los datos del dataset, los due√±os de los restaurantes pueden optimizar el control de insumos, reduciendo desperdicios y mejorando la rentabilidad.

3. identificaci√≥n de Tendencias de Consumo:
Este dataset ofrece una visi√≥n importante sobre qu√© tipo de tacos tienen mayor demanda, en qu√© zonas geogr√°ficas se consume m√°s y en qu√© d√≠as de la semana se concentran los pedidos. Esto ayuda a las marcas a realizar una planificaci√≥n adecuada de su producci√≥n y estrategia de marketing.

```python
# ==============================================================================
# GR√ÅFICO 2: DIAGRAMA DE FLUJO DEL PROCESO DE DATOS PARA ESTE NEGOCIO
#
# Este gr√°fico muestra el "viaje" de los datos, desde su captura
# hasta c√≥mo se convierten en una decisi√≥n estrat√©gica para elnegocio.
# ==============================================================================

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- Datos estructurados para el diagrama ---
data_pipeline = {
    'Captura del Dato': [
        'Ubicaci√≥n del cliente',
        'Tipo y tama√±o de taco',
        'Cantidad de toppings',
        'Tiempo de entrega',
        'Distancia recorrida',
        'Propina entregada',
        'Patrones de d√≠a'
    ],
    'An√°lisis Aplicado': [
        'Segmentaci√≥n Geogr√°fica',
        'An√°lisis de Preferencias',
        'Control de Costos',
        'Evaluaci√≥n de Eficiencia',
        'Optimizaci√≥n de Rutas',
        'Medici√≥n de Satisfacci√≥n',
        'An√°lisis Comparativo'
    ],
    'Decisi√≥n Estrat√©gica': [
        'Planificaci√≥n de Marketing',
        'Ingenier√≠a de Men√∫',
        'Ajuste de Precios',
        'Mejora de Operaciones',
        'Reducci√≥n de Gastos',
        'Programas de Lealtad',
        'Gesti√≥n de Demanda'
    ]
}

# --- Creaci√≥n de la visualizaci√≥n ---
fig, ax = plt.subplots(figsize=(20, 10))
fig.set_facecolor("#F8F9FA")

# Propiedades de las cajas y flechas
box_style = dict(boxstyle='round,pad=0.5', fc='#FFFFFF', ec='#ADB5BD', lw=1.5)
arrow_style = dict(arrowstyle='-|>', color='#6C757D', lw=2, mutation_scale=20)
y_pos = 0.5
x_positions = [0.15, 0.5, 0.85]
stage_titles = ['1. Captura del Dato', '2. An√°lisis Aplicado', '3. Decisi√≥n Estrat√©gica']
colors = ['#005A9C', '#0077B6', '#0096C7'] # Tonos de azul

for i, (stage_title, data_list) in enumerate(data_pipeline.items()):
    x = x_positions[i]

    # Unir la lista de datos en un solo bloque de texto
    text_content = '\n'.join(f'‚Ä¢ {item}' for item in data_list)

    # Dibujar la caja principal con el texto
    ax.text(x, y_pos, text_content, ha='center', va='center', fontsize=12, color='#495057',
            bbox=box_style, linespacing=1.6)

    # A√±adir el t√≠tulo del escenario
    ax.text(x, 0.9, stage_title, ha='center', va='center', fontsize=16,
            fontweight='bold', color=colors[i])

# Dibujar las flechas conectando las etapas
for i in range(len(x_positions) - 1):
    arrow = patches.FancyArrowPatch(
        (x_positions[i] + 0.15, y_pos),
        (x_positions[i+1] - 0.15, y_pos),
        connectionstyle="arc3,rad=0",
        **arrow_style
    )
    ax.add_patch(arrow)

# Formato final del gr√°fico
ax.axis('off')
ax.set_ylim(0, 1)
ax.set_xlim(0, 1)
plt.suptitle('Pipeline de Inteligencia de Negocio: Taco Sales', fontsize=24, fontweight='bold', color='#212529')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
